```python
import numpy as np
import seaborn as sns
```

```python
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
type(iris)
```

# 1. Quantitative plots

## 1.1. Histogram

```python
sns.displot(iris['sepal_length'])
```

```python
sns.displot(iris['sepal_length'], bins=5)
```

```python
sns.displot(x='sepal_length', hue='species', data=iris)
```

## 1.2. Rug plot

```python
sns.rugplot(data=iris, x='sepal_length', hue='species')
```

## 1.3. Scatterplot

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
```

## 1.4. Pairplot

```python
sns.pairplot(data=iris)
```

```python
sns.pairplot(iris, hue='species')
```

# 2. Categorical Plots

## 2.1. Bar plot

```python
sns.barplot(x='species', y='sepal_length', data=iris)
```

```python
sns.barplot(x='species', y='sepal_length', data=iris, estimator=np.std)
```

## 2.2. Boxplot

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
```

# 2.3. Heatmap

### 2.3.1. Heatmap for correlation table

```python
iris.corr() # compute pair-wise correlations of columns
```

```python
sns.heatmap(iris.corr(), annot=True, cmap='coolwarm')
```

### 2.3.2. Heatmap for pivot table

```python
flights = sns.load_dataset('flights')
```

```python
flights
```

```python
pivot_table = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
pivot_table
```

```python
sns.heatmap(pivot_table, cmap='coolwarm', linecolor='white', linewidth=1)
```

### 2.3.3. Clustermap

```python
sns.clustermap(pivot_table, cmap='coolwarm')
```

```python
sns.clustermap(pivot_table, cmap='coolwarm', standard_scale=1)
```
